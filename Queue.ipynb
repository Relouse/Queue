**Класс очереди**

Сравнение происходит следующим образом:

1) Если длины очередей не равны нулю, то сравнение происходит по длине, либо в случае равенства длин - по приоритету первого элемента

 2) Если длина одной из очередей равна нулю, то сравнивается по длине очереди 

```python
class MyDeque():  # Класс очереди
    def __init__(self,str):
        self.str = str  # Сама очередь
        
    def __gt__(self, other):  # Сравнение(знак больше)
        if (len(self.str) !=0) & (len(other.str) !=0):
            return (len(self.str) > len(other.str)) | ((len(self.str) == len(other.str)) & (self.str[0] < other.str[0]))
        else:
            return (len(self.str) > len(other.str))
        
    def __lt__(self, other):  # Сравнение(знак меньше)
        if (len(self.str) !=0) & (len(other.str) !=0):
            return (len(self.str) < len(other.str)) | ((len(self.str) == len(other.str)) & (self.str[0] > other.str[0]))
        else:
            return (len(self.str) < len(other.str))
        
    def __eq__(self, other):  # Сравнение(знак равенство) - равенство длин и первых элементов
            return (len(self.str) == len(other.str)) & (self.str[0] == other.str[0])
        
    def append(self, symbol):  # Добавление элемента в конец очереди
        self.str += symbol
        
    def put(self):  # Удаление первого элемента из очереди
        self.str = self.str[1:]
        
    def print_f(self):  # Вывод первого элемента очереди
        print(self.str[0])
```

Функция для разбора двух параллельных очередей согласно логике разбора

```python
def parse_deques(a,b):
    while (len(a.str) != 0) or (len(b.str) != 0):
            if a > b:
                a.print_f()
                a.put()
            else:
                b.print_f()
                b.put()
```

```python
a = MyDeque('AABBAB')
b = MyDeque('ABABAAAAXVXZ')
```

```python
a > b
```

```python
parse_deques(a,b)
```
